```python
import pandas as pd
import numpy as np
import json
```

```python
Refineddata=pd.read_csv('../data/Refined_dataset.csv')
uniquedata=Refineddata.drop_duplicates()
uniquedata.describe(include='all')
print(uniquedata.isnull().sum())
filleddata=uniquedata.copy()
filleddata['Age']=uniquedata['Age'].fillna(uniquedata['Age'].median())
filleddata['Embarked']=uniquedata['Embarked'].fillna('U')
print(filleddata.isnull().sum())
```

```python
null=filleddata.copy()
null['Cabin'] = filleddata['Cabin'].isna().astype(int)
correlation=null[['Cabin','Survived']].corr()
print(correlation)
filleddata['Cabin']=uniquedata['Cabin'].fillna('U1')
print(filleddata.isnull().sum())
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(sparse_output=False,drop=None)
array=encoder.fit_transform(filleddata[['Sex']])
names=encoder.get_feature_names_out(['Sex'])
encoded=pd.DataFrame(array,columns=names,index=filleddata.index)
encoded=pd.concat([filleddata.drop(['Sex'],axis=1),encoded],axis=1)
#print(encoded)

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
encoded['Cabin']=le.fit_transform(encoded['Cabin'])
#print(encoded)

encoder=OneHotEncoder(sparse_output=False,drop=None)
array=encoder.fit_transform(encoded[['Embarked']])
names=encoder.get_feature_names_out(['Embarked'])
encoded=pd.concat([encoded.drop(['Embarked'],axis=1),pd.DataFrame(array,columns=names,index=encoded.index)],axis=1)
#print(encoded)
```

```python
#print(encoded.dtypes)
column=encoded.select_dtypes(include=['int64','float64']).columns.tolist()
from sklearn.preprocessing import MinMaxScaler
encoded[column]=MinMaxScaler().fit_transform(encoded[column])
#print(encoded)
normalized=encoded.copy()
#print(normalized.describe())
```

```python
NewFeatures=encoded.columns.drop('Survived').to_list()
x=encoded[NewFeatures]
y=encoded['Survived']
x.to_csv('../data/newfeatures.csv',index=False)
y.to_csv('../data/newlabel.csv',index=False)
```

```python
x=pd.read_csv('../data/newfeatures.csv')
y=pd.read_csv('../data/newlabel.csv')
new_dataset_summary = {
    "shape":pd.concat([x,y],axis=1).shape,
    "columns":pd.concat([x,y],axis=1).columns.tolist(),
    "features": {
        "columns":x.columns.tolist(),
        "missing_values":x.isnull().sum().to_dict()
    },
    "target": {
        "column":y.columns.tolist(),
        "missing_values":y.isnull().sum().to_dict()
    },
    "data_types":pd.concat([x,y],axis=1).dtypes.apply(lambda x: x.name).to_dict(),
    "statistics":pd.concat([x,y],axis=1).describe().to_dict(),
    

}
output_file='../output/JuhayerAshfi_week1day3_data.json'
with open(output_file,'w') as f:
    json.dump(new_dataset_summary,f,indent=4)
```
