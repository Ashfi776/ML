```python
import pandas as pd
import numpy as np
dataframe= pd.read_csv('../data/train.csv')
```

```python
Features=['Pclass','Sex','Age','SibSp','Fare','Parch','Cabin','Embarked']
x=dataframe[Features]
y=dataframe['Survived']
x.to_csv('../data/features.csv',index=False)
y.to_csv('../data/label.csv',index=False)
```

```python
x=pd.read_csv('../data/features.csv')
y=pd.read_csv('../data/label.csv')
Refineddataset=pd.concat([x,y],axis=1)
Refineddataset.to_csv('../data/Refined_dataset.csv',index=False)
```

```python
import json
Refineddataset=pd.read_csv('../data/Refined_dataset.csv')
dataset_summary = {
    "shape":Refineddataset.shape,
    "columns":Refineddataset.columns.tolist(),
    "features": {
        "columns":x.columns.tolist(),
        "missing_values":x.isnull().sum().to_dict()
    },
    "target": {
        "column":y.columns.tolist(),
        "missing_values":y.isnull().sum().to_dict()
    },
}
output_file='../output/JuhayerAshfi_week1day1_data.json'
with open(output_file,'w') as f:
    json.dump(dataset_summary,f,indent=4)
```
