```python
Features=['Pclass','Sex','Age','SibSp','Fare','Parch','Cabin','Embarked']
x=dataframe[Features]
y=dataframe['Survived']
print("Features:")
print(x)
print("\n")
print("Label:")
print(y)
```

```python
x.to_csv('../data/features.csv',index=False)
y.to_csv('../data/label.csv',index=False)
```

```python
x=pd.read_csv('../data/features.csv')
y=pd.read_csv('../data/label.csv')
Refineddataset=pd.concat([x,y],axis=1)
Refineddataset.to_csv('../data/Refined_dataset.csv',index=False)
```
