```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import json
```

```python
Refineddataset=pd.read_csv('../data/Refined_dataset.csv')
NumericalSummary=Refineddataset.describe()
WholeSummary=Refineddataset.describe(include='all')
for col in Refineddataset:
    col_counts=Refineddataset[col].value_counts()
    col_percentages=(Refineddataset[col].value_counts(normalize=True))*100
correlation=(Refineddataset.drop(['Sex','Embarked','Cabin'],axis=1)).corr()
```

```python
for col in ['Age','Fare']:
    plt.figure()
    plt.hist(Refineddataset[col], bins=20)
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.savefig(f'../plots/{col}_histogram.png')
    plt.close()
for col in ['Pclass','Sex','SibSp','Parch','Cabin','Embarked','Survived']:
    plt.figure()
    plt.bar(Refineddataset[col].value_counts().index,Refineddataset[col].value_counts().values)
    plt.title(f'{col} count')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.savefig(f'../plots/{col}_bar.png')
    plt.close()
for col in ['Age','Fare']:
    plt.figure()
    sb.boxplot(y=Refineddataset[col],x=Refineddataset['Survived'])
    plt.title(f'{col} count vs Survived')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.savefig(f'../plots/{col}_box.png')
    plt.close()

plt.figure()
sb.heatmap(Refineddataset.drop(['Sex','Cabin','Embarked'],axis=1).corr(),annot=True)
plt.title('Correlation Heatmap')
plt.savefig('../plots/correlation_heatmap.png')
plt.close()
```

```python
plot_data=[]
for col in ['Age','Fare']:
    plot_data.append({
        "filename":f'../plots/{col}_histogram.png',
        "type":"histogram",
        "title":f'{col} Distribution',
        "xlabel":col,
        "ylabel":"Frequency"
    })
for col in ['Pclass','Sex','SibSp','Parch','Cabin','Embarked','Survived']:
    plot_data.append({
        "filename":f'../plots/{col}_bar.png',
        "type":"bar",
        "title":f'{col} Count',
        "xlabel":col,
        "ylabel":"Count"
    })
for col in ['Age','Fare']:
    plot_data.append({
        "filename":f'../plots/{col}_box.png',
        "type":"boxplot",
        "title":f'{col} vs Survived',
        "xlabel":"Survived",
        "ylabel":col
    })
plot_data.append({
    "filename":'../plots/correlation_heatmap.png',
    "type":"heatmap",
    "title":"Correlation Heatmap",
})
findings={
    "numerical_summary":NumericalSummary.to_dict(),
    "Whole_summary":WholeSummary.to_dict(),
    "Correlation":correlation.to_dict(),
    "Plots":plot_data
}
output_file='../output/JuhayerAshfi_week1day2_data.json'
with open(output_file,'w') as f:
    json.dump(findings,f,indent=4)
```
